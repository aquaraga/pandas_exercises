### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

%matplotlib notebook

sns.set(style="ticks", color_codes=True)
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/Visualization/Titanic_Desaster/train.csv). 

### Step 3. Assign it to a variable titanic 

```python
titanic = pd.read_csv('train.csv')
```

### Step 4. Set PassengerId as the index 

```python
titanic.set_index('PassengerId', inplace=True)
titanic.head()
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
plt.pie(titanic['Sex'].value_counts(), labels = titanic['Sex'].unique(), explode = (0.1, 0), autopct = '%1.1f%%');
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python

plt.figure()
femalepred = titanic.Sex == 'female'
females = titanic[femalepred]
males = titanic[-femalepred]
plt.scatter(females.Age, females.Fare, label = 'Female', edgecolors='black')
plt.scatter(males.Age, males.Fare, label = 'Male', edgecolors='black')
plt.legend()
plt.gca().set_title('Distribution of fare by age')
plt.gca().set_xlabel('Age (in yrs)')
plt.gca().set_ylabel('Fare (in unknown currency)');
```

### Step 7. How many people survived?

```python
titanic.Survived.sum()
```

### Step 8. Create a histogram with the Fare payed

```python
plt.figure()
plt.hist(titanic.Fare, bins = 20)
plt.gca().set_title('Distribution of fares')
plt.gca().set_xlabel('Fare')
plt.gca().set_ylabel('Number of passengers');
```
