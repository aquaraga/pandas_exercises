### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# set this so the 
%matplotlib inline
```

### Step 2. Create the DataFrame that should look like the one below.

```python
scores = pd.DataFrame({'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
                      'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
                      'age': [42, 52, 36, 24, 73],
                      'female': [0, 1, 1, 0, 1],
                      'preTestScore': [4, 24, 31, 2, 3],
                      'postTestScore': [25,94,57,62,70]})
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
plt.figure()

females = scores[scores['female'] == 1]
males = scores[scores['female'] == 0]
plt.scatter(females['preTestScore'], females['postTestScore'], s = females['age'], label = 'Females')
plt.scatter(males['preTestScore'], males['postTestScore'], s = males['age'], label = 'Males')
plt.legend()
plt.gca().set_xlabel('Pre-test score')
plt.gca().set_ylabel('Post-test score')
plt.gca().set_title('Pre-test vs Post-test scores, sized by age');
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
plt.figure()

females = scores[scores['female'] == 1]
males = scores[scores['female'] == 0]
plt.scatter(females['preTestScore'], females['postTestScore'], s = females['postTestScore'] * 4.5, label = 'Females')
plt.scatter(males['preTestScore'], males['postTestScore'], s = males['postTestScore'] * 4.5, label = 'Males')
plt.legend()
plt.gca().set_xlabel('Pre-test score')
plt.gca().set_ylabel('Post-test score')
plt.gca().set_title('Pre-test vs Post-test scores, sized by post-test score');
```
